```python
import joblib
import pandas as pd
import numpy as np

# Load model
model = joblib.load("cardiotoxicity_model.pkl")

# Feature order (must match training exactly)
feature_cols = [
    "age", "sex", "resting_hr", "systolic_bp", "diastolic_bp",
    "hrv_rmssd", "qtc_baseline", "baseline_lvef",
    "num_cycles", "dose_per_cycle", "cumulative_dose"
]

# Label names
label_map = {0: "Low Risk", 1: "Moderate Risk", 2: "High Risk"}


# ---- 7 sample patients ----
patients = pd.DataFrame([
    # very safe heart + low exposure → LOW
    [35, 0, 70, 118, 76, 42, 410, 65, 2, 40, 80],

    # mild stress + moderate exposure → MODERATE
    [52, 1, 85, 130, 84, 28, 445, 55, 4, 55, 220],

    # weak heart + high cumulative dose → HIGH
    [60, 1, 92, 142, 90, 18, 470, 45, 6, 70, 420],

    # young but aggressive dosing → MODERATE/HIGH boundary
    [40, 0, 88, 125, 82, 30, 455, 58, 5, 65, 300],

    # strong heart + many cycles but low dose → LOW/MODERATE
    [48, 1, 75, 122, 80, 36, 430, 62, 6, 35, 210],

    # elderly + borderline ECG + high exposure → HIGH
    [70, 0, 95, 150, 95, 15, 480, 42, 7, 75, 500],

    # average everything → MODERATE
    [55, 1, 82, 135, 85, 25, 450, 52, 5, 60, 300]
], columns=feature_cols)


# ---- Prediction ----
probs = model.predict_proba(patients)
preds = np.argmax(probs, axis=1)

for i in range(len(patients)):
    print(f"\nPatient {i+1}")
    print("Prediction:", label_map[preds[i]])
    print("Probabilities:", probs[i])
```
