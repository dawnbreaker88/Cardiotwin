```python
import numpy as np
import pandas as pd

np.random.seed(42)
N = 800

# -----------------------------
# 1. Baseline patient physiology
# -----------------------------
age = np.random.normal(55, 12, N).clip(25, 85)
sex = np.random.binomial(1, 0.5, N)

resting_hr = np.random.normal(72 + 0.1*(age-50), 9, N).clip(50, 115)

systolic_bp = np.random.normal(125 + 0.5*(age-50), 16, N).clip(90, 185)
diastolic_bp = np.random.normal(80 + 0.2*(age-50), 10, N).clip(55, 115)

hrv_rmssd = np.random.normal(
    40 - 0.3*(age-50) - 0.2*(resting_hr-70),
    12,
    N
).clip(5, 90)

qtc_baseline = np.random.normal(420 + 0.4*(age-50), 22, N).clip(360, 510)

baseline_lvef = np.random.normal(
    62 - 0.15*(age-50) - 0.05*(systolic_bp-120),
    6,
    N
).clip(35, 75)


# -----------------------------
# 2. Drug exposure
# -----------------------------
num_cycles = np.random.randint(3, 9, N)
dose_per_cycle = np.random.normal(60, 10, N).clip(30, 85)
cumulative_dose = num_cycles * dose_per_cycle


# -----------------------------
# 3. Hidden biological susceptibility
#    (model CANNOT see this)
# -----------------------------
latent_sensitivity = np.random.normal(1.0, 0.6, N).clip(0.2, 2.5)


# -----------------------------
# 4. Nonlinear stochastic dose response
# -----------------------------
dose_effect = 1 / (1 + np.exp(-(cumulative_dose - 350) / 70))

delta_lvef = (
    -10 * dose_effect * latent_sensitivity
    - 2 * latent_sensitivity
    + 0.02 * (baseline_lvef - 55)
    + 0.015 * (hrv_rmssd - 35)
    + np.random.normal(0, 4.5, N)   # strong biological noise
)

qtc_change = (
    25 * dose_effect * latent_sensitivity
    + np.random.normal(0, 9, N)
)

hrv_change = (
    -12 * dose_effect * latent_sensitivity
    + np.random.normal(0, 7, N)
)


# -----------------------------
# 5. Post-treatment noisy measurements
# -----------------------------
post_lvef = (baseline_lvef + delta_lvef + np.random.normal(0, 3, N)).clip(15, 75)
post_qtc = (qtc_baseline + qtc_change + np.random.normal(0, 8, N)).clip(360, 580)


# -----------------------------
# 6. Probabilistic cardiotoxic risk
#    (OVERLAPPING classes)
# -----------------------------
risk_score = (
    0.02 * (cumulative_dose - 320)
    - 0.05 * (post_lvef - 55)
    + 0.015 * (post_qtc - 460)
    + 0.8 * np.random.normal(0, 1, N)   # irreducible uncertainty
)

risk_prob = 1 / (1 + np.exp(-risk_score))

risk_label = pd.cut(
    risk_prob,
    bins=[0, 0.33, 0.66, 1],
    labels=["Low", "Moderate", "High"]
)


# -----------------------------
# 7. Final dataframe
# -----------------------------
df = pd.DataFrame({
    "age": age,
    "sex": sex,
    "resting_hr": resting_hr,
    "systolic_bp": systolic_bp,
    "diastolic_bp": diastolic_bp,
    "hrv_rmssd": hrv_rmssd,
    "qtc_baseline": qtc_baseline,
    "baseline_lvef": baseline_lvef,
    "num_cycles": num_cycles,
    "dose_per_cycle": dose_per_cycle,
    "cumulative_dose": cumulative_dose,
    "delta_lvef": delta_lvef,
    "qtc_change": qtc_change,
    "hrv_change": hrv_change,
    "post_lvef": post_lvef,
    "post_qtc": post_qtc,
    "risk_label": risk_label
})

df["risk_encoded"] = df["risk_label"].map({"Low": 0, "Moderate": 1, "High": 2})

df.head()
```

```python
df.to_csv("risk3.csv", index=False)
df.head()
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, ConfusionMatrixDisplay

df = pd.read_csv("risk3.csv")
print(df.shape)
df.head()
```

```python
X = df.drop(columns=[
    "risk_label",
    "risk_encoded",
    "delta_lvef",
    "qtc_change",
    "hrv_change",
    "post_lvef",
    "post_qtc"
])

y = df["risk_encoded"]

print(X.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
model = lgb.LGBMClassifier(
    n_estimators=400,
    learning_rate=0.04,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(
    model,
    X_train,
    y_train,
    cv=cv,
    scoring="roc_auc_ovr"
)

print("CV AUC scores:", scores)
print("Mean CV AUC:", scores.mean())
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.model_selection import train_test_split

X = df.drop("risk_encoded", axis=1)
y = df["risk_encoded"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)

# Class weights  force attention to moderate risk
class_weights = {0: 1.0, 1: 2.5, 2: 1.2}

model = LGBMClassifier(
    n_estimators=400,
    learning_rate=0.03,
    max_depth=4,
    num_leaves=20,
    class_weight=class_weights,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)

print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob, multi_class="ovr"))
```
