```python
# =========================
# 1. Imports
# =========================
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

from lightgbm import LGBMClassifier


# =========================
# 2. Prepare X and y
# =========================
# Make sure labels are clean and numeric
df["risk_label"] = df["risk_label"].str.strip().str.lower()

df["risk_encoded"] = df["risk_label"].map({
    "low": 0,
    "moderate": 1,
    "high": 2
})

df = df.dropna(subset=["risk_encoded"])
df["risk_encoded"] = df["risk_encoded"].astype(int)

# Features and target
X = df.drop(["risk_label", "risk_encoded"], axis=1)
y = df["risk_encoded"]


# =========================
# 3. Traintest split
# =========================
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    stratify=y,
    random_state=42
)


# =========================
# 4. Handle class imbalance
# =========================
class_weights = {
    0: 1.0,   # low risk
    1: 2.5,   # moderate risk  give more importance
    2: 1.2    # high risk
}


# =========================
# 5. Train LightGBM model
# =========================
model = LGBMClassifier(
    n_estimators=400,
    learning_rate=0.03,
    max_depth=4,
    num_leaves=20,
    class_weight=class_weights,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

model.fit(X_train, y_train)


# =========================
# 6. Predictions
# =========================
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)


# =========================
# 7. Evaluation
# =========================
print("\n=== Classification Report ===")
print(classification_report(y_test, y_pred))

print("\n=== Confusion Matrix ===")
print(confusion_matrix(y_test, y_pred))

auc = roc_auc_score(y_test, y_prob, multi_class="ovr")
print("\n=== ROC-AUC ===")
print(auc)
```

```python
# =============================
# 1. Imports
# =============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, ConfusionMatrixDisplay
from lightgbm import LGBMClassifier


# =============================
# 2. Prepare features and label
# =============================
# Clean label text
df["risk_label"] = df["risk_label"].str.strip().str.lower()

# Encode target
df["risk_encoded"] = df["risk_label"].map({
    "low": 0,
    "moderate": 1,
    "high": 2
})

# Drop rows with missing label
df = df.dropna(subset=["risk_encoded"])
df["risk_encoded"] = df["risk_encoded"].astype(int)


# --- ONLY baseline + exposure features ---
feature_cols = [
    "age","sex","resting_hr","systolic_bp","diastolic_bp",
    "hrv_rmssd","qtc_baseline","baseline_lvef",
    "num_cycles","dose_per_cycle","cumulative_dose"
]

X = df[feature_cols]
y = df["risk_encoded"]


# =============================
# 3. Train-test split
# =============================
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    stratify=y,
    random_state=42
)


# =============================
# 4. Handle class imbalance
# =============================
class_weights = {0: 1.0, 1: 2.5, 2: 1.2}


# =============================
# 5. Train LightGBM model
# =============================
model = LGBMClassifier(
    n_estimators=400,
    learning_rate=0.03,
    max_depth=4,
    num_leaves=20,
    class_weight=class_weights,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

model.fit(X_train, y_train)


# =============================
# 6. Predictions
# =============================
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)


# =============================
# 7. Evaluation
# =============================
print("\n=== Classification Report ===")
print(classification_report(y_test, y_pred))

print("\n=== Confusion Matrix ===")
cm = confusion_matrix(y_test, y_pred)
print(cm)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()

auc = roc_auc_score(y_test, y_prob, multi_class="ovr")
print("\n=== ROC-AUC ===")
print(auc)
```

```python
import numpy as np

def custom_predict(probabilities, threshold=0.6):
    preds = []

    for p in probabilities:
        low_p, mod_p, high_p = p

        # confident high
        if high_p > threshold:
            preds.append(2)

        # confident low
        elif low_p > threshold:
            preds.append(0)

        # otherwise  moderate (uncertain zone)
        else:
            preds.append(1)

    return np.array(preds)
```

```python
custom_preds = custom_predict(probs, threshold=0.6)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, custom_preds))
```

```python
import joblib
import numpy as np

model = joblib.load("cardiotoxicity_model.pkl")

label_map = {0: "Low Risk", 1: "Moderate Risk", 2: "High Risk"}

def predict_risk(patient_df):
    probs = model.predict_proba(patient_df)[0]

    # custom moderate logic
    if 0.35 < probs[1] < 0.65:
        pred_class = 1
    else:
        pred_class = np.argmax(probs)

    return label_map[pred_class], probs
```
